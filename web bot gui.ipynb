```python
from webbot import Browser
import time
import random
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import tkinter.filedialog
from tkinter.ttk import Progressbar
web = Browser()
```

```python
def open_browser(event):
    web.go_to('https://www.easports.com/fifa/ultimate-team/web-app/#')

def login(event):
    web.click('Login')
    
def credentials(event):
    web.type('****@****.**' , into = 'Email')
    web.type('*********', into = 'Password')
    web.click('Log In')
    web.click('Send Security Code')
    
def snipe(event):
    p = random.randint(30, 70)
    from progressbar import ProgressBar
    pbar = ProgressBar()
    progress_var = DoubleVar()
    progress = ttk.Progressbar(root, variable=progress_var, length = 300, maximum=(p-1))
    progress.grid(row = 4, columnspan = 3,padx = 50)
    for i in pbar(range(p)):
        url_then = web.get_current_url()
        k = i%2
        string = str(i)
        if len(string) == 1:
            number = '0'+string
        if len(string) > 1:
            number = string
    #print(number, end=", ")
        if k == 0:
            web.click(css_selector = '#ut-search-wrapper div div div div div div button.btn-standard.decrement-value', number = 1)
        if k == 1:
            web.click(css_selector = '#ut-search-wrapper div div div div div div button.btn-standard.increment-value', number = 1)
        web.click('Search')
    #time.sleep(random.uniform(0.05, 0.07))
        web.click('Buy Now', number = 1)
        web.click('OK')
        web.click('Send to Transfer List')
        web.click(classname = 'btn-navigation')
        url_now = web.get_current_url()
        if url_now != url_then:
            break
        progress_var.set(i)
        root.update()
        time.sleep(random.uniform(0.3, 1.0))
```

```python
root = Tk()
frame = Frame(root)
ttk.Style().theme_use('default')

open_chrome = ttk.Button(root, text = 'Go to FUT Web App')
open_chrome.bind('<Button-1>', open_browser)
open_chrome.grid(row = 0, column = 0)

click_login = ttk.Button(root, text = 'Login')
click_login.bind('<Button-1>', login)
click_login.grid(row = 1, column = 0)

email_pass = ttk.Button(root, text = 'Input Email and Password')
email_pass.bind('<Button-1>', credentials)
email_pass.grid(row = 2, column = 0)

snipe_but = ttk.Button(root, text = 'Start Sniping')
snipe_but.bind('<Button-1>', snipe)
snipe_but.grid(row = 3, column = 0)

root.mainloop()
```
